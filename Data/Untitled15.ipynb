```python
import pandas as pd

# 使用上傳的實際文件名稱替換 'uploaded_file_name.csv'
df = pd.read_csv('processed_data.csv')

# 檢查資料的前幾行
df.head()
```

```python
# 進行欄位值的轉換
df['Sex'] = df['Sex'].replace({'female': 0, 'male': 1})
df['Housing'] = df['Housing'].replace({'own': 0, 'rent': 1, 'free': 2})
df['Saving accounts'] = df['Saving accounts'].replace({
    'little': 0,
    'moderate': 1,
    'quite rich': 2,
    'rich': 3
})
df['Checking account'] = df['Checking account'].replace({
    'little': 0,
    'moderate': 1,
    'quite rich': 2,
    'rich': 3
})
df['Purpose'] = df['Purpose'].replace({
    'car': 0,
    'furniture/equipment': 1,
    'radio/TV': 2,
    'domestic appliances': 3,
    'repairs': 4,
    'education': 5,
    'business': 6,
    'vacation/others': 7
})

# 檢查轉換後的資料
print(df.head())
```

```python
df.to_csv('processed_data_converted.csv', index=False)
```
